**Step2 : Data Cleaning**

```python
data_bookings.describe()   # here we are seeing -ve data that is not possible for no_guests.
```

**Cleaning Invalid Guest**

```python
data_bookings[data_bookings.no_guests<=0]
```

**As you can see above, number of guests having less than zero value represents data error. We can ignore these records.**

```python
data_bookings = data_bookings[data_bookings.no_guests >= 0]
data_bookings.shape
```

**(2) Outlier removal in revenue generated**

```python
data_bookings.revenue_generated.min(), data_bookings.revenue_generated.max()
```

```python
data_bookings.revenue_generated.mean(), data_bookings.revenue_generated.std()
```

```python
avg,std = data_bookings.revenue_generated.mean(), data_bookings.revenue_generated.std()
```

```python
higher_limit = avg + 3*std
higher_limit
```

```python
lower_limit = avg - 3*std
lower_limit
```

```python
data_bookings[data_bookings.revenue_generated < 0]
# there is no -ve value so we are good.

```

```python

data_bookings[data_bookings.revenue_generated >= higher_limit]
```

**Step 3. Data Transformation**

Create occupancy percentage column

```python
data_agg.head(3)
```

```python
data_agg['occ_pct'] = data_agg.apply(lambda row: row['successful_bookings']/row['capacity'], axis=1)
```

```python
data_agg.head()
```

```python
data_agg['occ_pct'] = data_agg['occ_pct'].apply(lambda x: round(x*100, 2))
data_agg.head(3)
```

```python
data_agg.head()
```
